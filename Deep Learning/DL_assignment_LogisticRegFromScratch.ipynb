```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('winequality-red.csv',sep=';')
```

```python
data
```

```python
dataY = np.where(data["quality"]<=6,0,1)
dataY
```

```python
dataX = data.iloc[:,:-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(dataX,dataY,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
X_train= sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

Logistic Regression PART 2


```python

class MyLogisticRegression:
    def __init__(self, alp=0.02, it=1000):
        self.alp = alp
        self.it = it

    def fit(self, x, y):
        self.x = x
        self.y = y
        self.m, self.n = x.shape
        self.w = np.zeros(self.n)
        self.b = 0

        for i in range(self.it):
            self.upd()

    def upd(self):
        yh = 1 / (1 + np.exp(-(self.x.dot(self.w) + self.b)))
        dw = (1 / self.m) * np.dot(self.x.T, (yh - self.y))
        db = (1 / self.m) * np.sum(yh - self.y)

        self.w -= self.alp * dw
        self.b -= self.alp * db

    def predict(self, x):
        yp = 1 / (1 + np.exp(-(x.dot(self.w) + self.b)))
        yp = np.where(yp > 0.5, 1, 0)
        return yp
```

```python
model = MyLogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred,Y_test)
score*100
```
