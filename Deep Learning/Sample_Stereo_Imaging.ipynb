```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def compute_disparity_map(left_image_path, right_image_path):
    left_image = cv2.imread(left_image_path, 0)
    right_image = cv2.imread(right_image_path, 0)

    if left_image.shape != right_image.shape:
        height, width = left_image.shape
        right_image = cv2.resize(right_image, (width, height))

    stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)

    disparity = stereo.compute(left_image, right_image)
    disparity_normalized = cv2.normalize(disparity, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    plt.imshow(disparity_normalized)
    plt.show()

    return disparity_normalized

left_image_path = 'left_sample.png'
right_image_path = 'right_sample.png'

disparity_map = compute_disparity_map(left_image_path, right_image_path)
```
